```python
import os 
import shutil 
while(1): 
    ch=int(input("Enter choice:\n1.Get current working directory\n2.List all files and directories in the current directory\n3.Create a new directory\n4.Verify if the new directory has been created\n5.copy a file into this directory\n6.Exit\n")) 
    if ch==1: 
        print("\ncurrent working directory:",os.getcwd()) 
    elif ch==2: 
        print("\nlist of files and directories in current directory:") 
        for ele in os.listdir(os.getcwd()): 
            print(ele) 
    elif ch==3: 
        directory_name=input("enter directory name:")
        os.mkdir(directory_name)
        print("New directory created sucessfully*")
    elif ch==4: 
        print("Verify if the new directory has been created:")
        if os.path.exists(directory_name):
            print("Directory created successfully*")
        else:
            print("Directory not created*")
    elif ch==5: 
        print("copy a file into this directory:")
        copy_file=input("enter file to copy into directory:")
        if shutil.copy(copy_file, directory_name) :
            print("copied a file into directory")
        else:
            print("cannot copied a file")
    elif ch==6:
        break
    else:
        print("invalid choice*")
```

```python
def count_file(file_path):
    f1=open(file_path,'r')
    line_count=0
    word_count=0
    for line in f1:
        line_count+=1
        for data in line:
             if data==' 'or data=='\n':
                word_count+=1
    print("Line count=",line_count)
    print("Word count=",word_count)
file="speech_barack_obama.txt"
print("Barack Obama file:")
count_file(file)
file="speech_michelle_obama.txt"
print("Michelle Obama file:")
count_file(file)
file="speech_donald_trump.txt"
print("Donald_trump file:")
count_file(file)
file="speech_melina_trump.txt"
print("Melina_trump file:")
count_file(file)
```

```python
class Teacher:
    def __init__(self, name, age, subject):
        self.name = name
        self.age = age
        self.subject = subject

    def display_info(self):
        print("Name:",self.name," \nAge:",self.age,"\nSubject:", self.subject)


class Professor(Teacher):
    def __init__(self, name, age, subject, research_area):
        super().__init__(name, age, subject)
        self.research_area = research_area

    def display_info(self):
        super().display_info()
        print("Research Area:",self.research_area)


class AstProfessor(Teacher):
    def __init__(self, name, age, subject, num_publications):
        super().__init__(name, age, subject)
        self.num_publications = num_publications

    def display_info(self):
        super().display_info()
        print("Number of Publications: ",self.num_publications)


class AsProfessor(Teacher):
    def __init__(self, name, age, subject,years_experience):
        super().__init__(name, age, subject)
        self.years_experience = years_experience

    def display_info(self):
        super().display_info()
        print("Years of Experience:",self.years_experience)


# Create objects and display their information
p = Professor("Dr.Alice", 50, "Physics", "Quantum Mechanics")
ap = AstProfessor("Bob", 40, "Astronomy", 100)
asp = AsProfessor("Charlie", 45, "Computer Science", 20)
print("Professor Details:")
p.display_info()
print("Assistant Professor Details:")
ap.display_info()
print("Associate Professor Details:")
asp.display_info()
```

```python
def calculate_sales_commission(sales_per_week):
    total_sales = sum(sales_per_week)
    commission_rate = 5/100  # 5% commission rate
    commission = 0

    if total_sales >= 50000:
        commission = total_sales * commission_rate

    # Determine remarks based on total sales
    if total_sales >= 80000:
        remarks = "Excellent"
    elif 60000 <= total_sales < 80000:
        remarks = "Good"
    elif 40000 <= total_sales < 60000:
        remarks = "Average"
    else:
        remarks = "Work Hard"

    return total_sales, commission, remarks

num_salesmen = int(input("Enter the number of salesmen: "))
salesmen_data = []

for i in range(num_salesmen):
    salesman_name = input(f"Enter name of salesman {i + 1}: ")
    sales_per_week = [int(input(f"Enter sales for week {j + 1}: ")) for j in range(4)]  # Assuming 4 weeks in a month

    total_sales, commission, remarks = calculate_sales_commission(sales_per_week)

        # Store data in a list of tuples
    salesmen_data.append((salesman_name, total_sales, commission, remarks))

    # Display results
print("\nSalesman Data:")
for data in salesmen_data:
    print("Salesman: ",data[0]," Total Sales: ",data[1], "Commission: ",data[2], "Remarks:", data[3])
```

```python
def copy_alternative_lines(file1, file2):
    try:
        with open(file1, 'r') as file_1, open(file2, 'w') as file_2:
            lines = file_1.readlines()
            # Use list slicing to get odd-numbered lines
            odd_lines = lines[::2]
            file_2.writelines(odd_lines)
        print("Alternative lines copied from file1 to file2 successfully\n")
    except:
        print("Error: File file1 not found.")
    else:
        print("Odd numbered lines copied successfully*\n")
        with open(file2,'r')as f:
            for line in f:
                print(line)
# Example usage:
file1 = 'file1.txt'  
file2 = 'file2.txt'
copy_alternative_lines(file1, file2)
```

```python
import numpy as np
import pylab
import matplotlib.pyplot as plt
def plot_curves():
    # Generate data points
    x = np.linspace(0,20,1000)
    
    # Sine curve
    y_sine = np.sin(x)
    
    # Cosine curve
    y_cosine = np.cos(x)
    
    plt.plot(x, y_sine, label='Sine')
    plt.plot(x, y_cosine, label='cosine')
    plt.legend(loc='upper right')
    
    y_polynomial = x**2 - 3*x + 1
    y_exponential=np.exp(x)
    
    plt.plot(x, y_polynomial, label='Ploynomial')
    plt.plot(x, y_exponential, label='Exponential')
    plt.legend(loc='upper right')
    pylab.ylim(-1.5, 5)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Sine,Cosine,Ploynomial,Exponential Graph")
plot_curves()
```

```python
from multipledispatch import dispatch
class example:
    @dispatch(int,int)
    def add(self,a,b):
        return a+b
    @dispatch(int,int,int)
    def add(self,a,b,c):
        return a+b+c
    @dispatch(float,float,float)
    def add(self,a,b,c):
        return a+b+c
    @dispatch(float,float)
    def add(self,a,b):
        return a+b
    @dispatch(int,float)
    def add(self,a,b):
        return a+b
    @dispatch(str,str)
    def add(self,a,b):
        return a+b
obj=example()
print(obj.add(20,10))
print(obj.add(40,6,4))
print(obj.add(6.5,1.0,7.7))
print(obj.add(50.5,6.4))
print(obj.add(20,7.9))
print(obj.add("Hello*","Welcome"))
```
