```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure()

# The .gca() function is "Get Current Axes".
# It returns the current axes matching the **kwargs
 # If the axes doesn't exist (it doesn't here), it'll be created automatically
 # Here, we're packing the returned value into an `ax` object
 # You can create 3D plots like this if you aren't using subplots
ax = fig.gca(projection='3d')
X = np.arange(-10, 10, 0.40)
Y = np.arange(-10, 10, 0.40)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
 # Creating the surface with Numpy variables
 # And specifying the Colormap to be Spectral
surf = ax.plot_surface(X, Y, Z, cmap=cm.Spectral)

 # Adding a colorbar to the right of the plot
 # For a user-friendly bar of values
fig.colorbar(surf, shrink=0.5,aspect=5)
plt.show()
```
