```python
lst=[]
n=int(input("enter size of list:"))
for i in range(n):
    x=int(input("enter elements to the list:"))
    lst.append(x)
largest=secondlargest=lst[0]
for i in lst:
    if i>largest:
        secondlargest=largest
        largest=i
        #when first element is second largest
    elif i>secondlargest and i<largest:
        secondlargest=i
        #when first element is secondlargest
    elif secondlargest==largest:
        secondlargest=i
print("List:",lst)
print("second largest element:",secondlargest)
```

```python
n=int(input("enter number:"))
sumofseries=1
for i in range(1,n+1):
    fact=1
    for j in range(1,i+1):
        fact=fact*j
    sumofseries=sumofseries+1/fact
print("Sum of series:",sumofseries)
```

```python
def addition(num1,num2):
    return num1+num2
def substraction( num1, num2):
    return num1-num2
def multiplication( num1, num2):
    return num1*num2
def division( num1, num2):
    return num1/num2
while(1):
    print("Select choice:")
    print("1.Addition\n2.Substraction\n3.Multiplication\n4.Division\n5.Exit")
    ch=int(input("enter choice:"))
    if ch==1:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Addition of two number:",addition(n1,n2))
    elif ch==2:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Substraction of two numbers:",substraction(n1,n2))
    elif ch==3:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Product of two numbers:",multiplication(n1,n2))
    elif ch==4:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("quotient of two numbers:",division(n1,n2))
    elif ch==5:
        break
    else:
        print("invalid choice select correct choice")
```

```python
import math
def areaofcircle(r):
    return math.pi*r*r
def areaoftriangle(b,h):
    return 0.5*b*h
def areaofrectangle(w,l):
    return w*l
def perimeterofcircle(r):
    return 2*math.pi*r
def perimeteroftriangle(s1,b,s2):
    return s1+b+s2
def perimeterofrectangle(w,l):
    return 2*(l+w)
while(1):
    ch=int(input("enter choice:\n1.circle\n2.Triangle\n3.Rectangle\n4.Exit\n"))
    if ch==1:
        r=float(input("enter radius of circle:"))
        print("\nArea of circle= ",areaofcircle(r),"\n")
        print("Perimeter of circle: ",perimeterofcircle(r),"\n")
    elif ch==2: 
        b=float(input("enter breadth: "))
        h=float(input("enter height: "))
        print("\nArea of Triangle= ",areaoftriangle(b,h),"\n")
        side1=float(input("enter side1 of triangle:"))
        side2=float(input("enter side2 of triangle:"))
        base=float(input("enter base of triangle:"))
        print("Perimeter of Triangle: ",perimeteroftriangle(side1,base,side2),"\n")
    elif ch==3:
        w=float(input("enter width of rectangle:"))
        l=float(input("enter length of rectangle:"))
        print("\nArea of rectangle= ",areaofrectangle(w,l),"\n")
        print("Perimeter of Rectangle: ",perimeterofrectangle(l,w),"\n")
    elif ch==4:
        break
    else:
        print("invalid choice:")
    
```
