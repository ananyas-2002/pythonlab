```python
import csv
import json
import xml
while(1):
    ch=int(input("enter choice: \n1.CSV file\n2.JSON file\n3.XML file\n4.exit\n"))
    if ch==1:
        # Reading CSV
        with open('data.csv', 'r') as file:
            reader = csv.reader(file)
            for row in reader:
                print(row)

        # Writing CSV
        data_to_write = [['Name', 'Age', 'City'],
                         ['John', '25', 'New York'],
                         ['Alice', '30', 'San Francisco']]

        with open('output.csv', 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data_to_write)
    elif ch==2:
        # Reading JSON
        with open('jsonfile.json', 'r') as file:
            data = json.load(file)
            print(data)

        # Writing JSON
        data_to_write_json = {
                    "people": [
                    {"name": "John", "age": 25, "city": "New York"},
                    {"name": "Alice", "age": 30, "city": "San Francisco"}
                    ]
                    }

        with open('output.json', 'w') as file:
            json.dump(data_to_write_json, file, indent=2)
    elif ch==3:
        from bs4 import BeautifulSoup
        with open('../Documents/sample.xml','r')as f:
            data=f.read()
        Bs_data=BeautifulSoup(data,"xml")
        #print(Bs_data)
        b_student=Bs_data.find_all("student")
        print(b_student)
    elif ch==4:
        break
```
