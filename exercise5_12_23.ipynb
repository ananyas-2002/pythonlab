```python
class Employee:
    def __init__(self,empid,empname):
        self.empid=empid
        self.empname=empname
    def Empdetails(self):
        print("Employee id",self.empid)
        print("EMployee name:",self.empname)
class person(Employee):
    def __init__(self,empid,empname,department,salary):
        super(person,self).__init__(empid,empname)
        self.department=department
        self.salary=salary
    def Empdetails(self):
        print("Employeeid=",self.empid)
        print("Name=",self.empname)
        print("Department=",self.department)
        print("Salary=",self.salary)
class manager(person):
    def __init__(self,empid,empname,department,salary,managerid,managername):
        super(manager,self).__init__(empid,empname,department,salary)
        self.managerid=managerid
        self.managername=managername
    def Empdetails(self):
        print("Employeeid=",self.empid)
        print("Name=",self.empname)
        print("Department=",self.department)
        print("Salary=",self.salary)
        print("Manager Id=",self.managerid)
        print("Manager Name=",self.managername)
person1=person(1002,"Jhon","Tester",25000)
print("*******************Person class details****************************")
person1.Empdetails()
emp1=Employee(1003,"Danial")
print("*******************Employee class details****************************")
emp1.Empdetails()
mng1=manager(1002,"Jhon","Tester",25000,20202,"samson")
print("*******************Manager class details****************************")
mng1.Empdetails()
```

```python
class student:
    def __init__(self,name,registernumber,course,marks):
        self.name=name
        self.registernumber=registernumber
        self.course=course
        self.marks=marks
    def showStudentdetails(self):
        print("----------------------------------------------------------------------------------")
        print("Name:",end="\t\t")
        print("RegNo:",end="\t\t")
        print("Course:",end="\t\t")
        print("Marks:")
        print("----------------------------------------------------------------------------------")
        print(self.name,end="\t\t")
        print(self.registernumber,end="\t\t")
        print(self.course,end="\t\t")
        print(self.marks)
        print("----------------------------------------------------------------------------------")
    def Total(self):
        print("Total mark:",sum(marks))
    def Average(self):
        total=sum(marks)
        average=total/5
        print("Average mark=",average)
    def grade(self):
        average=sum(marks)/5
        if average>=80 and average<=100:
            print("A grade")
        elif average>=50 and average<=70:
            print("B grade")
        elif average>=30 and average<=60:
                print("C grade")
        elif average>=10 and average<=20:
            print("E grade")
        else:
            print("Failed")
name=input("enter name of student:")
regnumber=int(input("enter register number:"))
course=input("enter course:")
marks=[]
for i in range(5):
    marks.append(int(input("enter mark:")))
st=student(name,regnumber,course,marks)
print("*************Student details**********************")
st.showStudentdetails()
st.Total()
st.Average()
st.grade()
```

```python
class Employee:
    def __init__(self,empid,empname,designation,salary):
        self.empid=empid
        self.empname=empname
        self.designation=designation
        self.salary=salary
    def empGross(self):
        DA=5/100*salary
        HRA=10/100*salary
        Gross=salary+DA+HRA
        return Gross
    def empNet(self):
        LIC=3/100*salary
        PF=8/100*salary
        deduction=LIC+PF
        Netsalary=self.empGross()-deduction
        return Netsalary
    def Empdetails(self):
        print("----------------------------------------------------------------------------------")
        print("EmpId",end="\t\t")
        print("EmpName:",end="\t\t")
        print("Designation:",end="\t\t")
        print("Salary:")
        print("----------------------------------------------------------------------------------")
        print(self.empid,end="\t\t")
        print(self.empname,end="\t\t\t")
        print(self.designation,end="\t\t")
        print(self.salary)
        print("----------------------------------------------------------------------------------")
for i in range(5):
    empid.append(int(input("enter employee id:")))
    name.append(input("enter name of employee:"))
    designation.append(input("enter designation of employee:"))
    salary.append(float(input("enter basic salary:")))
print("************Employee Details************************")
emp1=Employee(empid[0],name[0],designation[0],salary[0])
emp2=Employee(empid[1],name[1],designation[1],salary[1])
emp3=Employee(empid[2],name[2],designation[2],salary[2])
emp4=Employee(empid[3],name[3],designation[3],salary[3])
emp5=Employee(empid[4],name[4],designation[4],salary[4])
emp1.Empdetails()
emp2.Empdetails()
emp3.Empdetails()
emp4.Empdetails()
emp5.Empdetails()
print("Gross Salary" ,emp1.empGross())
print("Net Salary",emp1.empNet())
```
