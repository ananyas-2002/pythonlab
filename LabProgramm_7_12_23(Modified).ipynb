```python
class Employee:
    def __init__(self,empid,empname,designation,salary):
        self.empid=empid
        self.empname=empname
        self.designation=designation
        self.salary=salary
    def empGross(self):
        DA=5/100*self.salary
        HRA=10/100*self.salary
        Gross=self.salary+DA+HRA
        return Gross
    def empNet(self):
        LIC=3/100*self.salary
        PF=8/100*self.salary
        deduction=LIC+PF
        Netsalary=self.empGross()-deduction
        return Netsalary
    def Empdetails(self):
        print(self.empid,end="\t\t")
        print(self.empname,end="\t\t\t")
        print(self.designation,end="\t\t")
        print(self.salary,end="\t\t")
        print(self.empGross(),end="\t\t")
        print(self.empNet())
        print("---------------------------------------------------------------------------------------------------------------")
empid=[]
empname=[]
designation=[]
salary=[]
for i in range(5):
    empid.append(int(input("enter employee id:")))
    empname.append(input("enter name of employee:"))
    designation.append(input("enter designation of employee:"))
    salary.append(float(input("enter basic salary:")))
print("************Employee Details************************")
emp1=Employee(empid[0],empname[0],designation[0],salary[0])
emp2=Employee(empid[1],empname[1],designation[1],salary[1])
emp3=Employee(empid[2],empname[2],designation[2],salary[2])
emp4=Employee(empid[3],empname[3],designation[3],salary[3])
emp5=Employee(empid[4],empname[4],designation[4],salary[4])
print("------------------------------------------------------------------------------------------------------------------------")
print("EmpId:",end="\t\t")
print("EmpName:",end="\t\t")
print("Designation:",end="\t\t")
print("Salary:",end="\t\t")
print("Gross:",end="\t\t")
print("Net:")
print("------------------------------------------------------------------------------------------------------------------------")
emp1.Empdetails()
emp2.Empdetails()
emp3.Empdetails()
emp4.Empdetails()
emp5.Empdetails()
```
