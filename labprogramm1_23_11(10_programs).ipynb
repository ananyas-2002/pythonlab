```python
lst=[]
n=int(input("enter size of list:"))
for i in range(n):
    x=int(input("enter elements to the list:"))
    lst.append(x)
largest=secondlargest=lst[0]
for i in lst:
    if i>largest:
        secondlargest=largest
        largest=i
        #when first element is second largest
    elif i>secondlargest and i<largest:
        secondlargest=i
        #when first element is secondlargest
    elif secondlargest==largest:
        secondlargest=i
print("List:",lst)
print("second largest element:",secondlargest)
```

```python
def addition(num1,num2):
    return num1+num2
def substraction( num1, num2):
    return num1-num2
def multiplication( num1, num2):
    return num1*num2
def division( num1, num2):
    return num1/num2
while(1):
    print("Select choice:")
    print("1.Addition\n2.Substraction\n3.Multiplication\n4.Division\n5.Exit")
    ch=int(input("enter choice:"))
    if ch==1:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Addition of two number:",addition(n1,n2))
    elif ch==2:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Substraction of two numbers:",substraction(n1,n2))
    elif ch==3:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("Product of two numbers:",multiplication(n1,n2))
    elif ch==4:
        n1=float(input("enter first number:"))
        n2=float(input("enter second number:"))
        print("quotient of two numbers:",division(n1,n2))
    elif ch==5:
        break
    else:
        print("invalid choice select correct choice")
```

```python
import math
def areaofcircle(r):
    return math.pi*r*r
def areaoftriangle(b,h):
    return 0.5*b*h
def areaofrectangle(w,l):
    return w*l
def perimeterofcircle(r):
    return 2*math.pi*r
def perimeteroftriangle(s1,b,s2):
    return s1+b+s2
def perimeterofrectangle(w,l):
    return 2*(l+w)
while(1):
    ch=int(input("enter choice:\n1.circle\n2.Triangle\n3.Rectangle\n4.Exit\n"))
    if ch==1:
        r=float(input("enter radius of circle:"))
        print("\nArea of circle= ",areaofcircle(r),"\n")
        print("Perimeter of circle: ",perimeterofcircle(r),"\n")
    elif ch==2: 
        b=float(input("enter breadth: "))
        h=float(input("enter height: "))
        print("\nArea of Triangle= ",areaoftriangle(b,h),"\n")
        side1=float(input("enter side1 of triangle:"))
        side2=float(input("enter side2 of triangle:"))
        base=float(input("enter base of triangle:"))
        print("Perimeter of Triangle: ",perimeteroftriangle(side1,base,side2),"\n")
    elif ch==3:
        w=float(input("enter width of rectangle:"))
        l=float(input("enter length of rectangle:"))
        print("\nArea of rectangle= ",areaofrectangle(w,l),"\n")
        print("Perimeter of Rectangle: ",perimeterofrectangle(l,w),"\n")
    elif ch==4:
        break
    else:
        print("invalid choice:")
    
```

```python
import math
def equationroot(a,b,c):
    descriminator=b*b-4*a*c
    if descriminator>0:
        squarevalue=math.sqrt(descriminator)
        print("roots are real and diferent:")
        root1=(-b+squarevalue)/(2*a)
        root2=(-b-squarevalue)/(2*a)
        print("Root1=",root1)
        print("Root2=",root2)
    elif descriminator==0:
        print("roots are real and same roots:")
        root=-b/2*a
        print("Root=",root)
    else:
        squarevalue=math.sqrt(abs(descriminator))
        print("roots are complex:")
        print("Root1:",-b/(2*a),"+",squarevalue,"i")
        print("Root2:",-b/(2*a),"-",squarevalue,"i")
a=float(input("enter a value:"))
b=float(input("enter b value:"))
c=float(input("enter c value:"))
if a==0:
    print("input correct quadratic values:")
else:
    equationroot(a,b,c)
```

```python
class Employee:
    def __init__(self,empid,empname,designation,salary):
        self.empid=empid
        self.empname=empname
        self.designation=designation
        self.salary=salary
    def empGross(self):
        DA=5/100*self.salary
        HRA=10/100*self.salary
        Gross=self.salary+DA+HRA
        return Gross
    def empNet(self):
        LIC=3/100*self.salary
        PF=8/100*self.salary
        deduction=LIC+PF
        Netsalary=self.empGross()-deduction
        return Netsalary
    def Empdetails(self):
        print(self.empid,end="\t\t")
        print(self.empname,end="\t\t\t")
        print(self.designation,end="\t\t")
        print(self.salary,end="\t\t")
        print(self.empGross(),end="\t\t")
        print(self.empNet())
        print("---------------------------------------------------------------------------------------------------------------")
empid=[]
empname=[]
designation=[]
salary=[]
for i in range(5):
    empid.append(int(input("enter employee id:")))
    empname.append(input("enter name of employee:"))
    designation.append(input("enter designation of employee:"))
    salary.append(float(input("enter basic salary:")))
print("************Employee Details************************")
emp1=Employee(empid[0],empname[0],designation[0],salary[0])
emp2=Employee(empid[1],empname[1],designation[1],salary[1])
emp3=Employee(empid[2],empname[2],designation[2],salary[2])
emp4=Employee(empid[3],empname[3],designation[3],salary[3])
emp5=Employee(empid[4],empname[4],designation[4],salary[4])
print("------------------------------------------------------------------------------------------------------------------------")
print("EmpId:",end="\t\t")
print("EmpName:",end="\t\t")
print("Designation:",end="\t\t")
print("Salary:",end="\t\t")
print("Gross:",end="\t\t")
print("Net:")
print("------------------------------------------------------------------------------------------------------------------------")
emp1.Empdetails()
emp2.Empdetails()
emp3.Empdetails()
emp4.Empdetails()
emp5.Empdetails()
```
