```python
def copy_alternative_lines(file1, file2):
    try:
        with open(file1, 'r') as file_1, open(file2, 'w') as file_2:
            lines = file_1.readlines()
            # Use list slicing to get odd-numbered lines
            odd_lines = lines[::2]
            file_2.writelines(odd_lines)
        print("Alternative lines copied from file1 to file2 successfully\n")
    except:
        print("Error: File file1 not found.")
    else:
        print("Odd numbered lines copied successfully*\n")
        with open(file2,'r')as f:
            for line in f:
                print(line)
# Example usage:
file1 = 'file1.txt'  
file2 = 'file2.txt'
copy_alternative_lines(file1, file2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def plot_curves():
    # Generate data points
    x = np.linspace(0,20,1000)

    # Sine curve
    y_sine = np.sin(x)

    # Cosine curve
    y_cosine = np.cos(x)

    # Polynomial curve (example: quadratic)
    y_polynomial = x**2 + 2*x + 1

    # Exponential curve
    y_exponential = np.exp(x)

    # Plotting
    plt.figure(figsize=(10, 8))

    plt.subplot(2, 2, 1)
    plt.plot(x, y_sine, label='Sine')
    plt.title('Sine Curve')
    plt.legend(loc='upper right')

    plt.subplot(2, 2, 2)
    plt.plot(x, y_cosine, label='Cosine')
    plt.title('Cosine Curve')
    plt.legend(loc='upper right')

    plt.subplot(2, 2, 3)
    plt.plot(x, y_polynomial, label='Polynomial')
    plt.title('Polynomial Curve')
    plt.legend(loc='upper left')

    plt.subplot(2, 2, 4)
    plt.plot(x, y_exponential, label='Exponential')
    plt.title('Exponential Curve')
    plt.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

# Call the function to plot the curves
plot_curves()
```

```python
import numpy as np
import pylab
import matplotlib.pyplot as plt
def plot_curves():
    # Generate data points
    x = np.linspace(0,20,1000)
    
    # Sine curve
    y_sine = np.sin(x)
    
    # Cosine curve
    y_cosine = np.cos(x)
    
    plt.plot(x, y_sine, label='Sine')
    plt.plot(x, y_cosine, label='cosine')
    plt.legend(loc='upper right')
    
    y_polynomial = x**2 - 3*x + 1
    y_exponential=np.exp(x)
    
    plt.plot(x, y_polynomial, label='Ploynomial')
    plt.plot(x, y_exponential, label='Exponential')
    plt.legend(loc='upper right')
    pylab.ylim(-1.5, 5)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Sine,Cosine,Ploynomial,Exponential Graph")
plot_curves()
```
